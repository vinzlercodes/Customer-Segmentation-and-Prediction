## Complete Pre-Processing Data Function (summary of all above cells)
This is mostly for the supervised machine learning part of the project so that the above cells written right now for **Clustering**, dont have to be repeated for SML

```python
def data_preprocess(data, cluster, dataname=None):
    if cluster:
        if dataname == 'azdias':
            data = data[data.isnull().sum(axis=1) <= 16].reset_index(drop=True)
        elif dataname == 'customers':            
            data.drop(columns=['CUSTOMER_GROUP', 'ONLINE_PURCHASE', 'PRODUCT_GROUP'], inplace=True)
    print("Dropping columns with more than 65% missing values")      
    columns_drop = ['EXTSEL992', 'KK_KUNDENTYP', 'ALTER_KIND1', 'ALTER_KIND2', 'ALTER_KIND3', 'ALTER_KIND4']
    
    data = data.drop(columns_drop,axis=1)
    data = data.drop(['D19_LETZTER_KAUF_BRANCHE'],axis=1)
    data = data.drop(['EINGEFUEGT_AM'],axis=1)
    
    print('columns dropped')
    
    # computing the correlation matrix and dropping the correlated matrix
    print('Computing Correlation matrix')
    correlation_matrix = data.corr().abs()
    upper_data = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(np.bool))
    drop_column  = [col for col in upper_data.columns if any(upper_data[col] > .7)]
    data = data.drop(drop_column, axis=1)
    print('shape after dropping the correlated columns: {}'.format(data.shape))
    
    #filling the missing and unknown values with -1 
    print('filling nan vlaues with -1')
    data[['CAMEO_INTL_2015', 'CAMEO_DEUG_2015']] = data[['CAMEO_INTL_2015', 'CAMEO_DEUG_2015']].replace(['X','XX'],-1)
    data[['CAMEO_INTL_2015', 'CAMEO_DEUG_2015']] = data[['CAMEO_INTL_2015', 'CAMEO_DEUG_2015']].fillna(-1)
    data[['CAMEO_INTL_2015', 'CAMEO_DEUG_2015']] = data[['CAMEO_INTL_2015', 'CAMEO_DEUG_2015']].astype(int)
    data[['OST_WEST_KZ', 'CAMEO_DEU_2015']] = data[['OST_WEST_KZ', 'CAMEO_DEU_2015']].fillna(-1)
    
     # filling na fields with 9 that has marked 9 as unknown
    data[data.columns[(data==9).any()]] = data[data.columns[(data==9).any()]].fillna(9)

    # filling na fields with 0 that has marked 0 as unknown
    data[data.columns[(data==0).any()]] = data[data.columns[(data==0).any()]].fillna(0)

    # filling na fields with -1 that has marked -1 as unknown
    data[data.columns[(data==-1).any()]] = data[data.columns[(data==-1).any()]].fillna(-1)
    
    print('values filled')
    #creating dummy variable for the categorical variables
    print('creating dummy variables')
    data = pd.get_dummies(data)
    print("Shape of data after creating dummy vairables: {}".format(data.shape))
    
    columns_data = list(data.columns.values)
    
    #imputing the nan value with mode
    print('Imputing nan values with mode')
    mode_impute = Imputer(missing_values='NaN',strategy='most_frequent',axis=0)
    data = mode_impute.fit_transform(data)
    data = pd.DataFrame(data)
    print('shape after imputing the values', data.shape)
    
    #converting all values in df to int
    data = data.astype(int)
    
    #reomving outliers
    print('removing outliers')
    if cluster:
        print('Removing outliers')
        data = data[(np.abs(stats.zscore(data)) < 6).all(axis=1)] 
        
    print('Shape before normalising: {}'.format(data.shape))
    normaliser = StandardScaler(copy=False).fit_transform(data)
    data = pd.DataFrame(normaliser,columns= columns_data)
    print('Shape after normalising: {}'.format(data.shape))
        
    data = data.set_index('LNR')
    return data
```

**We first reduce the dimension of our dataset further. We can now use PCA for dimensional reduction. To decide the exact components we want to keep, we can see the variance across the features. We want our dataset to maintain a high vairance for better results**

```python
#plotting the graph for pca
data_pca = PCA().fit(azdias)
plt.figure(figsize=(18,9))
plt.plot(np.cumsum(data_pca.explained_variance_ratio_))
plt.xlabel('Number of features')
plt.title("Variance distribution across features")
plt.ylabel('Cumulative Variance')
plt.show()
```

**We can see from the graph above, that at 220 components we are getting high variance which is neccessary for keeping critical information about the data. So now we will use PCA to reduce the dimension of data to 220.**

```python
def dimension_reduce(data, n_component=220):
    data_pca = PCA(n_components=n_component).fit(data)
    dimension_reduce = data_pca.transform(data)
    dimension_reduce = pd.DataFrame(dimension_reduce)
    print("Variance of reduced dataset: {}".format(data_pca.explained_variance_ratio_.sum()))
    return dimension_reduce
```

```python
#performind data reduction on the dataset
azdias_pca = dimension_reduce(azdias)
customers_pca = dimension_reduce(customers)
```

# Clustering

```python
#defining the no. of clusters
def k_mean_scores(X, k):
    k_means = KMeans(k)
    k_means_model = k_means.fit(X)
    return abs(k_means_model.score(X))
```

```python
# checking the scores for different no. of clusters
accuracies = []

for i in range(1, 25):
    accuracies.append(k_mean_scores(azdias_pca.sample(20000), i))
```

```python
marks = np.linspace(1,24,24)
plt.figure(figsize=(16,8))
plt.plot(marks, accuracies, linestyle='-', marker='x', color='blue')
plt.xticks(list(range(1,25,2)))
plt.xlabel('Number of Clusters')
plt.ylabel('Average Within-Cluster Distances')
plt.title(" Distance within clusters vs No. of clusters")
```

**As we can see from the above chart, that the average distance within the clusters almost flatterns when the number of clusters reaches 16. So we will use 16 clusters for our analysis**

```python
#making the model with 16 clusters
k_means = KMeans(n_clusters = 16)
k_means_model = k_means.fit(azdias_pca)
azdias_predict = k_means_model.predict(azdias_pca)
```

```python
#saving the prediction of the model in a pickle file

cluster_azdias = pd.DataFrame(azdias_predict, columns = ['Cluster'])
cluster_azdias.to_pickle('azdias_clustered')
```
